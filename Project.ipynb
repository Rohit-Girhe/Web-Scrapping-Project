```python
import requests
from bs4 import BeautifulSoup
import json

# Use the CORRECT URL with /multiple endpoint
url = "https://www.nobroker.in/property/commercial/rent/pune/multiple"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

print("="*70)
print("NOBROKER /MULTIPLE ENDPOINT DIAGNOSTIC")
print("="*70)

print(f"\nFetching: {url}")
response = requests.get(url, headers=headers, timeout=15)

print(f"Status Code: {response.status_code}")
print(f"Content Length: {len(response.content)} bytes")

soup = BeautifulSoup(response.content, 'html.parser')

print("\n" + "="*70)
print("ANALYSIS 1: SEARCH FOR KNOWN ELEMENTS")
print("="*70)

# Search for your known elements
titles = soup.find_all('h2', class_='flex items-center m-0 heading-6 font-semi-bold')
rents = soup.find_all('div', id='minimumRent')
locations = soup.find_all('div', class_=lambda c: c and 'text-gray-light' in ' '.join(c) if c else False)
deposits = soup.find_all('div', id='roomType')
builtups = soup.find_all('div', id='unitCode')

print(f"Titles (h2): {len(titles)}")
print(f"Rents (id='minimumRent'): {len(rents)}")
print(f"Locations (class='text-gray-light'): {len(locations)}")
print(f"Deposits (id='roomType'): {len(deposits)}")
print(f"Builtups (id='unitCode'): {len(builtups)}")

if titles:
    print(f"\nSample title: {titles[0].text.strip()[:80]}")

print("\n" + "="*70)
print("ANALYSIS 2: ALL DIVS WITH IDS")
print("="*70)

all_divs_with_ids = soup.find_all('div', id=True)
print(f"Total divs with ID attribute: {len(all_divs_with_ids)}")

if all_divs_with_ids:
    print("\nFirst 20 IDs found:")
    for i, div in enumerate(all_divs_with_ids[:20], 1):
        div_id = div.get('id', '')
        classes = ' '.join(div.get('class', []))[:60]
        print(f"  {i}. ID: {div_id}")
        if classes:
            print(f"     Class: {classes}")

print("\n" + "="*70)
print("ANALYSIS 3: SCRIPT TAGS WITH JSON")
print("="*70)

scripts = soup.find_all('script', type='application/json')
print(f"Script tags with type='application/json': {len(scripts)}")

if scripts:
    print(f"\nFound {len(scripts)} JSON scripts!")
    for i, script in enumerate(scripts[:3], 1):
        print(f"\n  Script {i} (first 300 chars):")
        print(f"  {script.string[:300]}...")

print("\n" + "="*70)
print("ANALYSIS 4: ALL SCRIPT TAGS")
print("="*70)

all_scripts = soup.find_all('script')
print(f"Total script tags: {len(all_scripts)}")

text_scripts = [s for s in all_scripts if s.string and len(s.string) > 100]
print(f"Scripts with significant content: {len(text_scripts)}")

for i, script in enumerate(text_scripts[:3], 1):
    content = script.string[:200]
    if any(x in content for x in ['property', 'rent', 'listing', 'properties']):
        print(f"\n  Text Script {i} (contains property data):")
        print(f"  {content[:150]}...")

print("\n" + "="*70)
print("ANALYSIS 5: COMMON CLASS PATTERNS")
print("="*70)

from collections import Counter

all_divs = soup.find_all('div', class_=True)
class_counter = Counter()

for div in all_divs:
    classes = ' '.join(div.get('class', []))
    if classes:
        class_counter[classes] += 1

print("Classes appearing 2-15 times (potential card patterns):")
for class_name, count in class_counter.most_common(50):
    if 1 < count <= 15:
        preview = class_name[:90]
        print(f"  Count: {count:2d} | {preview}")

print("\n" + "="*70)
print("ANALYSIS 6: CONTAINER PATTERNS")
print("="*70)

# Look for containers that might hold property cards
potential_containers = []

for div in all_divs_with_ids:
    # Check children for property elements
    has_title = div.find('h2', class_=lambda c: c and 'heading-6' in ' '.join(c) if c else False)
    has_rent = div.find('div', id='minimumRent')
    has_location = div.find('div', class_=lambda c: c and 'text-gray-light' in ' '.join(c) if c else False)

    fields_found = sum([bool(has_title), bool(has_rent), bool(has_location)])

    if fields_found >= 2:
        potential_containers.append({
            'id': div.get('id'),
            'fields': fields_found,
            'classes': ' '.join(div.get('class', []))[:80]
        })

print(f"Potential property containers found: {len(potential_containers)}")

if potential_containers:
    print("\nContainers with 2+ property fields:")
    for container in potential_containers[:5]:
        print(f"  ID: {container['id']}")
        print(f"  Fields: {container['fields']}")
        print(f"  Classes: {container['classes']}")

print("\n" + "="*70)
print("SUMMARY")
print("="*70)

total_elements = len(titles) + len(rents) + len(locations)

if total_elements > 0:
    print(f"✓ FOUND {total_elements} property elements!")
    print(f"  The page IS rendering content with static HTML")
    print(f"  Titles: {len(titles)}, Rents: {len(rents)}, Locations: {len(locations)}")
else:
    print(f"✗ NO property elements found in static HTML")
    print(f"\nPossible causes:")
    print(f"  1. Content is loaded via JavaScript (AJAX)")
    print(f"  2. Data is in JSON embedded in script tags")
    print(f"  3. Page structure is completely different")

if scripts:
    print(f"\n✓ Found {len(scripts)} JSON scripts - data might be embedded!")

print("\n" + "="*70)
```
