```python
# nobroker_requests_final_cleaned_v2.py
"""
Final cleaned scraper (v2) for NoBroker commercial rent listings (Pune sample).
- Stronger mojibake repair logic in clean_text() to remove leftover 'â', 'Â', 'Ã' artifacts.
- Strict sqft/parking extractors as before.
- Saves CSV with utf-8-sig for Excel compatibility.
"""

import requests
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
from urllib.parse import urljoin
import unicodedata

# ---------- Configuration ----------
BASE = "https://www.nobroker.in"
SEARCH_URL = ("https://www.nobroker.in/property/commercial/rent/pune/multiple?"
              "searchParam=W3sibGF0IjoxOC41NjQyNDUyLCJsb24iOjczLjc3Njg1MTEsInBsYWNlSWQiOiJDaElKeTlOZDhNLS13anNSZmF0Xy01Y1NrYUUiLCJwbGFjZU5hbWUiOiJCYW5lciJ9LHsibGF0IjoxOC41OTEyNzE2LCJsb24iOjczLjczODkwODk5OTk5OTk5LCJwbGFjZUlkIjoiQ2hJSjd4c0VTTUM3d2pzUjVkN0R3MXJyeWRBIiwicGxhY2VOYW1lIjoiSGluamF3YWRpIn0seyJsYXQiOjE4LjU1MzgyNDEsImxvbiI6NzMuOTQ3NjY4OSwicGxhY2VJZCI6IkNoSUpsYVNMS01QRHdqc1JTZ0JqT21FejZEZyIsInBsYWNlTmFtZSI6IktoYXJhZGkifV0=&radius=2.0&city=pune&locality=Baner,Hinjawadi,Kharadi")
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36'
}
REQUEST_SLEEP = 0.6
OUTPUT_CSV = "nobroker_requests_final_cleaned_v2.csv"

session = requests.Session()
session.headers.update(HEADERS)

# ---------- Text cleaning (improved) ----------
def _score_mojibake(s):
    """Lower is better: count of suspicious mojibake characters."""
    return (s.count('Ã') + s.count('Â') + s.count('â') + s.count('�'))

def clean_text(s):
    """
    Clean and normalize scraped text with multiple repair attempts:
    1. Remove zero-width chars.
    2. Try multiple re-decode repairs (latin1->utf-8, windows-1252->utf-8, utf-8->latin1).
    3. Pick the candidate with the least mojibake markers or that contains '₹'.
    4. Apply targeted replacements for common sequences.
    5. Unicode-normalize and collapse whitespace.
    """
    if s is None:
        return None
    s = str(s)

    # quick strip of invisible chars
    s = s.replace('\u200b', '').replace('\ufeff', '').strip()

    # prepare candidates list (start with original)
    candidates = [s]

    # attempt latin1 -> utf-8
    try:
        c = s.encode('latin1').decode('utf-8')
        candidates.append(c)
    except Exception:
        pass

    # attempt windows-1252 -> utf-8
    try:
        c2 = s.encode('cp1252').decode('utf-8')
        candidates.append(c2)
    except Exception:
        pass

    # attempt utf-8 -> latin1 (rare cases)
    try:
        c3 = s.encode('utf-8').decode('latin1')
        candidates.append(c3)
    except Exception:
        pass

    # Also try double-repair latin1->utf-8 -> latin1->utf-8
    try:
        temp = s.encode('latin1').decode('utf-8')
        c4 = temp.encode('latin1').decode('utf-8')
        candidates.append(c4)
    except Exception:
        pass

    # Evaluate candidates: pick one with least mojibake signs OR contains rupee
    best = s
    best_score = _score_mojibake(s)
    for cand in candidates:
        sc = _score_mojibake(cand)
        # prefer candidate containing rupee sign
        if '₹' in cand and '₹' in best:
            # tie-breaker by score
            if sc < best_score:
                best, best_score = cand, sc
        elif '₹' in cand and '₹' not in best:
            best, best_score = cand, sc
        else:
            if sc < best_score:
                best, best_score = cand, sc

    s = best

    # targeted replacements for known mojibake patterns
    repl_map = {
        'Ã¢Â‚Â¹': '₹',
        'â‚¹': '₹',
        'â\x82¹': '₹',
        '\xc2\xa0': ' ',
        '\u00a0': ' ',
        'Ã¢Â€Â“': '-',
        'Ã¢Â€Âœ': '"',
        'Ã¢Â€Â�': '"',
        'â€”': '-',
        'â€“': '-',
        'â€™': "'",
        'â€œ': '"',
        'â€\x9d': '"',
    }
    for k, v in repl_map.items():
        if k in s:
            s = s.replace(k, v)

    # If there's an 'â' immediately before digits (common for ₹), turn into rupee
    s = re.sub(r'â(?=\s*\d)', '₹', s)
    # If there is an isolated 'Â' or 'Ã' residual, remove them when they are stray
    s = re.sub(r'[ÂÃ]+(?=\s|$)', '', s)
    # remove any remaining replacement-character glyphs
    s = s.replace('�', '')

    # Unicode normalize
    try:
        s = unicodedata.normalize('NFKC', s)
    except Exception:
        pass

    # Remove control chars
    s = re.sub(r'[\x00-\x1f\x7f-\x9f]', '', s)

    # Collapse whitespace
    s = re.sub(r'\s+', ' ', s).strip()

    return s if s != '' else None

# ---------- Utility helpers ----------
def safe_text(tag):
    return tag.get_text(" ", strip=True) if tag else None

def parse_amount_rupee(text):
    if not text:
        return None
    t = str(text).replace('\u200b','').replace(',', '').replace('₹','').strip()
    if not t:
        return None
    m = re.search(r'([\d.]+)\s*(lacs?|lakh|lakhs|L|k|K)?', t, re.I)
    if not m or not m.group(1):
        return None
    try:
        val = float(m.group(1).strip())
    except Exception:
        return None
    unit = (m.group(2) or '').lower()
    if 'lac' in unit or 'lakh' in unit or unit == 'l':
        return int(val * 100000)
    if unit == 'k':
        return int(val * 1000)
    return int(val)

def parse_sqft(text):
    if not text:
        return None
    txt = str(text)
    m = re.search(r'([\d,]+(?:\.\d+)?)\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)\b', txt, re.I)
    if m and m.group(1):
        num = m.group(1).replace(',', '').strip()
        try:
            return float(num)
        except:
            return None
    return None

# ---------- Parking extractor ----------
_VALID_PARKING_WORDS = {
    'public','covered','covered parking','open','yes','no','no parking',
    'basement','valet','private','street','common','none','open parking'
}
_VALID_PARKING_RE = re.compile(r'^(?:' + r'|'.join(re.escape(w) for w in _VALID_PARKING_WORDS) + r')$', re.I)

def nearby_text_candidates(node, max_tokens=8):
    out = []
    parent = node.parent if hasattr(node, 'parent') else None
    if parent:
        for child in parent.find_all(recursive=False):
            txt = safe_text(child)
            if txt:
                out.append(txt.strip())
    try:
        prev_sib = node.previous_sibling
        next_sib = node.next_sibling
        for s in (prev_sib, next_sib):
            if s:
                if getattr(s, 'get_text', None):
                    txt = safe_text(s)
                    if txt:
                        out.append(txt.strip())
                else:
                    val = str(s).strip()
                    if val:
                        out.append(val)
    except Exception:
        pass
    if parent:
        for s in parent.stripped_strings:
            out.append(s.strip())
            if len(out) >= max_tokens:
                break
    seen = set()
    result = []
    for t in out:
        if not t:
            continue
        if t in seen:
            continue
        seen.add(t)
        result.append(t)
    return result[:max_tokens]

def is_good_parking_candidate(s):
    if not s:
        return False
    s = s.strip()
    if re.search(r'\d', s):
        return False
    if '₹' in s or 'rs' in s.lower():
        return False
    if 'sq' in s.lower() or 'per' in s.lower():
        return False
    normalized = re.sub(r'[^A-Za-z\s\-]', '', s).strip().lower()
    if not normalized:
        return False
    if _VALID_PARKING_RE.match(normalized):
        return True
    if len(normalized) <= 22 and re.match(r'^[a-z\s\-]+$', normalized):
        if any(k in normalized for k in ['park','valet','basement','covered','public','street','open','private','common','none']):
            return True
    return False

def extract_parking(card):
    label_nodes = card.find_all(string=re.compile(r'\bParking\b', re.I))
    for node in label_nodes:
        candidates = nearby_text_candidates(node)
        for cand in candidates:
            if is_good_parking_candidate(cand):
                return cand.strip()
        parent = node.parent
        if parent:
            for attr in ('title','aria-label'):
                v = parent.get(attr)
                if v and is_good_parking_candidate(v):
                    return v.strip()
    text = " ".join(card.stripped_strings)
    m = re.search(r'\b(Public|Covered|Valet|Basement|Open|Private|Street|Common|No Parking|No)\b', text, re.I)
    if m:
        cand = m.group(1)
        if is_good_parking_candidate(cand):
            return cand.strip()
    return None

# ---------- Price per sqft extractor ----------
def extract_price_per_sqft_from_card(card):
    def get_numeric_before_per_sq(s):
        if not s:
            return None
        s = s.replace('\u200b',' ').strip()
        m = re.search(r'([\d,.]+)\s*(?=(?:per|\b/|\bper\b).*(?:sq|Sq|SQ))', s, re.I)
        if m and m.group(1):
            try:
                return float(m.group(1).replace(',', ''))
            except:
                return None
        m2 = re.search(r'([\d,.]+)', s)
        if m2 and m2.group(1) and 'sq' in s.lower():
            try:
                return float(m2.group(1).replace(',', ''))
            except:
                return None
        return None
    for tag in card.find_all(['div','span']):
        cl = tag.get('class') or []
        if any('bg-grayDark' in c for c in cl):
            val = safe_text(tag)
            n = get_numeric_before_per_sq(val)
            if n:
                return n
    for tag in card.find_all(['div','span']):
        txt = safe_text(tag) or ''
        if 'per' in txt.lower() and 'sq' in txt.lower():
            n = get_numeric_before_per_sq(txt)
            if n:
                return n
    return None

# ---------- Furnishing ----------
def extract_furnishing(card):
    for s in card.stripped_strings:
        if re.search(r'\b(Unfurnished|Furnished|Semi[- ]Furnished|Semi[- ]Fitted)\b', s, re.I):
            m = re.search(r'\b(Unfurnished|Furnished|Semi[- ]Furnished|Semi[- ]Fitted)\b', s, re.I)
            if m:
                return m.group(1)
    nodes = card.find_all(string=re.compile(r'\bFurnishing\b', re.I))
    for node in nodes:
        parent = node.parent
        if parent:
            candidate = parent.find(lambda t: t.name in ('div','span') and t.get('class') and 'font-semi-bold' in ' '.join(t.get('class')))
            if candidate:
                v = safe_text(candidate) or ''
                if '₹' in v or 'sq' in v.lower() or re.search(r'\d{3,}', v):
                    continue
                if re.match(r'^[A-Za-z\s\-]{2,30}$', v.strip()):
                    return v.strip()
    return None

# ---------- Sqft extraction ----------
def extract_sqft_near_label(card):
    label = card.find(lambda t: t.name in ('div','span') and re.search(r'Built[-\s]?up|Builtup|Built Up|Builtup', (t.get_text(" ",strip=True) or ''), re.I))
    if label:
        parent = label.parent
        if parent:
            val = parent.find(lambda t: t.name in ('div','span') and t.get('class') and 'font-semi-bold' in ' '.join(t.get('class')))
            if val:
                txt = safe_text(val) or ''
                if re.search(r'\d.+(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)', txt, re.I):
                    return txt
            ptxt = safe_text(parent) or ''
            m = re.search(r'(\d{2,6}(?:\.\d+)?)\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)', ptxt or '', re.I)
            if m:
                return m.group(0)
    for t in card.find_all(['div','span','p','li']):
        txt = safe_text(t) or ''
        if re.search(r'\d{2,6}(?:\.\d+)?\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)', txt, re.I):
            return re.search(r'\d{2,6}(?:\.\d+)?\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)', txt, re.I).group(0)
    return None

# ---------- Rent & deposit ----------
def extract_rent_from_card(card):
    for s in card.stripped_strings:
        txt = s.strip()
        if '₹' in txt:
            return txt
    vals = [safe_text(x) for x in card.find_all(lambda t: t.name == 'div' and t.get('class') and 'font-semi-bold' in ' '.join(t.get('class')))]
    for v in vals:
        if v and ('₹' in v or re.search(r'\d', v)):
            return v
    return None

def extract_deposit(card):
    dep = None
    nodes = card.find_all(string=re.compile(r'\bDeposit\b', re.I))
    for node in nodes:
        parent = node.parent
        if parent:
            candidate = parent.find(lambda t: t.name in ('div','span') and t.get('class') and 'font-semi-bold' in ' '.join(t.get('class')))
            if candidate:
                dep = safe_text(candidate)
                break
    if dep:
        return dep
    vals = [safe_text(x) for x in card.find_all(lambda t: t.name == 'div' and t.get('class') and 'font-semi-bold' in ' '.join(t.get('class')))]
    if len(vals) >= 2:
        return vals[1]
    return None

# ---------- Available from & property type ----------
def extract_available_from(card):
    for s in card.stripped_strings:
        txt = s.strip()
        if re.match(r'\d{4}-\d{2}-\d{2}', txt):
            return txt
    nodes = card.find_all(string=re.compile(r'Available\s*From|Available from|Available\b', re.I))
    for node in nodes:
        parent = node.parent
        if parent:
            candidate = parent.find(lambda t: t.name in ('div','span') and t.get('class') and 'font-semi-bold' in ' '.join(t.get('class')))
            if candidate:
                return safe_text(candidate)
    return None

def extract_property_type(card):
    txt = " ".join(card.stripped_strings)
    m = re.search(r'\b(Shop|Office Space|Office|Retail|Warehouse|Showroom|Commercial|Restaurant)\b', txt, re.I)
    return m.group(1) if m else None

# ---------- Detail page parser ----------
def parse_detail_page(detail_url):
    if not detail_url:
        return {}
    try:
        r = session.get(detail_url, timeout=12)
        time.sleep(REQUEST_SLEEP)
    except Exception:
        return {}
    if r.status_code != 200:
        return {}
    dsoup = BeautifulSoup(r.text, 'html.parser')
    out = {}
    text = dsoup.get_text(" ", strip=True)
    for tag in dsoup.find_all("script", {"type":"application/ld+json"}):
        try:
            import json
            obj = json.loads(tag.string or "{}")
            if isinstance(obj, dict):
                if 'price' in obj and not out.get('Rent_detail'):
                    out['Rent_detail'] = obj.get('price')
                if 'address' in obj and not out.get('Location_detail'):
                    addr = obj.get('address')
                    if isinstance(addr, dict):
                        out['Location_detail'] = addr.get('streetAddress') or addr.get('addressLocality')
                    elif isinstance(addr, str):
                        out['Location_detail'] = addr
        except Exception:
            pass
    m_sq = re.search(r'([\d,]+(?:\.\d+)?)\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m)', text, re.I)
    if m_sq:
        out['Sqft_detail'] = m_sq.group(0)
    m_dep = re.search(r'([₹\d,.\sKkLl]+)\s*(?:Deposit|deposit)', text)
    if m_dep:
        out['Deposit_detail'] = m_dep.group(1)
    m_furn = re.search(r'\b(Unfurnished|Furnished|Semi[- ]Furnished)\b', text, re.I)
    if m_furn:
        out['Furnishing_detail'] = m_furn.group(1)
    m_av = re.search(r'Available\s*From[:\s]*([A-Za-z0-9\-\s]+)', text, re.I)
    if m_av:
        out['Available_from_detail'] = m_av.group(1).strip()
    m_park = re.search(r'\b(Public|Covered|No Parking|Yes|No|Basement|Valet|Private)\b\s*(?:Parking)', text, re.I)
    if m_park:
        out['Parking_detail'] = m_park.group(1)
    return out

# ---------- Main page parsing ----------
def parse_search_page(html, follow_details=False, detail_limit=None):
    soup = BeautifulSoup(html, 'html.parser')
    cards = soup.find_all('article')
    rows = []
    for idx, card in enumerate(cards, start=1):
        title_tag = card.find('h2', attrs={'class': lambda c: c and 'heading-6' in c})
        title = safe_text(title_tag) or safe_text(card.find(['h2','h3','h4']))

        loc_tag = card.find(lambda t: t.name in ('div','span') and t.get('class') and 'overflow-ellipsis' in " ".join(t.get('class')) and 'whitespace-nowrap' in " ".join(t.get('class')))
        location = safe_text(loc_tag) if loc_tag else None
        if location and title and title.strip() in location:
            location = location.replace(title, '').strip(' ,-')

        rent_raw = extract_rent_from_card(card)
        deposit_raw = extract_deposit(card)
        sqft_raw = extract_sqft_near_label(card)
        sqft_val = parse_sqft(sqft_raw) if sqft_raw else None

        price_pps = extract_price_per_sqft_from_card(card)
        rent_inr = parse_amount_rupee(rent_raw) if rent_raw else None
        if not price_pps and rent_inr and sqft_val and sqft_val > 0:
            try:
                price_pps = round(rent_inr / sqft_val, 2)
            except Exception:
                price_pps = None

        parking = extract_parking(card)
        furnishing = extract_furnishing(card)
        available_from = extract_available_from(card)
        property_type = extract_property_type(card)

        a = card.find('a', href=True)
        detail_link = urljoin(BASE, a.get('href')) if a else None

        detail_data = {}
        if follow_details:
            if detail_limit is not None and idx > detail_limit:
                detail_data = {}
            else:
                detail_data = parse_detail_page(detail_link)

        # prefer detail values if present
        if detail_data.get('Location_detail'):
            location = detail_data.get('Location_detail')
        if detail_data.get('Rent_detail'):
            rent_raw = detail_data.get('Rent_detail')
            rent_inr = parse_amount_rupee(rent_raw)
        if detail_data.get('Deposit_detail'):
            deposit_raw = detail_data.get('Deposit_detail')
        if detail_data.get('Sqft_detail'):
            sqft_raw = detail_data.get('Sqft_detail')
            sqft_val = parse_sqft(sqft_raw)
        if detail_data.get('Furnishing_detail'):
            furnishing = detail_data.get('Furnishing_detail')
        if detail_data.get('Available_from_detail'):
            available_from = detail_data.get('Available_from_detail')
        if detail_data.get('Parking_detail'):
            parking = detail_data.get('Parking_detail')

        # Clean text fields BEFORE storing
        title_c = clean_text(title)
        location_c = clean_text(location)
        rent_raw_c = clean_text(rent_raw)
        deposit_raw_c = clean_text(deposit_raw)
        sqft_raw_c = clean_text(sqft_raw)
        parking_c = clean_text(parking)
        furnishing_c = clean_text(furnishing)
        available_from_c = clean_text(available_from)
        property_type_c = clean_text(property_type)

        row = {
            "Title": title_c,
            "Location": location_c,
            "Rent_raw": rent_raw_c,
            "Rent_inr": rent_inr,
            "Deposit_raw": deposit_raw_c,
            "Sqft_raw": sqft_raw_c,
            "Sqft": sqft_val,
            "Price_per_sqft": price_pps,
            "Parking": parking_c,
            "Furnishing": furnishing_c,
            "Available_from": available_from_c,
            "Property_type": property_type_c,
            "Detail_URL": detail_link
        }
        rows.append(row)
    return rows

# ---------- Runner ----------
def run(save_csv=OUTPUT_CSV, follow_details=False, detail_limit=None):
    print("Requesting search page...")
    r = session.get(SEARCH_URL, timeout=15)
    r.raise_for_status()
    time.sleep(REQUEST_SLEEP)
    rows = parse_search_page(r.text, follow_details=follow_details, detail_limit=detail_limit)
    df = pd.DataFrame(rows)

    # Final cleanup on DataFrame text columns (extra safety)
    for col in ['Deposit_raw', 'Rent_raw', 'Location', 'Title', 'Parking', 'Furnishing', 'Sqft_raw']:
        if col in df.columns:
            df[col] = df[col].astype(str).apply(clean_text)

    # Save with BOM for better Excel support
    df.to_csv(save_csv, index=False, encoding='utf-8-sig')
    print(f"Saved {save_csv} — rows: {len(df)}")
    print(df.info())
    print(df.head(10).to_string(index=False))
    return df

if __name__ == "__main__":
    # By default don't follow detail pages (faster). Use follow_details=True to follow.
    df = run(save_csv=OUTPUT_CSV, follow_details=False, detail_limit=None)
```

```python
# nobroker_pune_commercial_scraper.py
"""
Scraper for NoBroker commercial rent listings across multiple Pune localities.
- Uses explicit locality URLs provided.
- Strong text cleaning (mojibake repair).
- Extracts title, location, rent, deposit, sqft, price_per_sqft, parking, furnishing, etc.
- Loops over all localities and writes/appends to a single CSV.
"""

import requests
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
from urllib.parse import urljoin
import unicodedata
import os

# ---------- Localities to scrape (name, FULL WORKING URL) ----------
locality_links = [
    ("Kothrud",
     "https://www.nobroker.in/property/commercial/rent/pune/Kothrud?searchParam=W3sibGF0IjoxOC41MDczNTE0LCJsb24iOjczLjgwNzY1NDMsInBsYWNlSWQiOiJDaElKbllTdk1yZV93anNSOEVULXMwaUxCOVEiLCJwbGFjZU5hbWUiOiJLb3RocnVkIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Kothrud&isMetro=false"),
    ("Baner",
     "https://www.nobroker.in/property/commercial/rent/pune/Baner?searchParam=W3sibGF0IjoxOC41NjQyNDUyLCJsb24iOjczLjc3Njg1MTEsInBsYWNlSWQiOiJDaElKeTlOZDhNLS13anNSZmF0Xy01Y1NrYUUiLCJwbGFjZU5hbWUiOiJCYW5lciIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Baner&isMetro=false"),
    ("Hinjawadi",
     "https://www.nobroker.in/property/commercial/rent/pune/Hinjawadi?searchParam=W3sibGF0IjoxOC41OTEyNzE2LCJsb24iOjczLjczODkwODk5OTk5OTk5LCJwbGFjZUlkIjoiQ2hJSjd4c0VTTUM3d2pzUjVkN0R3MXJyeWRBIiwicGxhY2VOYW1lIjoiSGluamF3YWRpIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Hinjawadi&isMetro=false"),
    ("Kharadi",
     "https://www.nobroker.in/property/commercial/rent/pune/Kharadi?searchParam=W3sibGF0IjoxOC41NTM4MjQxLCJsb24iOjczLjk0NzY2ODksInBsYWNlSWQiOiJDaElKbGFTTEtNUER3anNSU2dCak9tRXo2RGciLCJwbGFjZU5hbWUiOiJLaGFyYWRpIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Kharadi&isMetro=false"),
    ("Shivajiaagar",
     "https://www.nobroker.in/property/commercial/rent/pune/Shivajinagar?searchParam=W3sibGF0IjoxOC41MzE0NDE5LCJsb24iOjczLjg0NDU2MTcsInBsYWNlSWQiOiJDaElKZDFGNEhYbkF3anNSeUQxMHlvRnEyQ0EiLCJwbGFjZU5hbWUiOiJTaGl2YWppbmFnYXIiLCJzaG93TWFwIjpmYWxzZX1d&radius=2.0&city=pune&locality=Shivajinagar&isMetro=false"),
    ("Hadapsar",
     "https://www.nobroker.in/property/commercial/rent/pune/Hadapsar?searchParam=W3sibGF0IjoxOC41MDg5MzQsImxvbiI6NzMuOTI1OTEwMTk5OTk5OTksInBsYWNlSWQiOiJDaElKNmFyeGdmX3B3anNSVTR1c1ZUVTBZQ1UiLCJwbGFjZU5hbWUiOiJIYWRhcHNhciIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Hadapsar&isMetro=false"),
    ("Pimpri Chinchwad",
     "https://www.nobroker.in/property/commercial/rent/pune/Pimpri-Chinchwad?searchParam=W3sibGF0IjoxOC42MjYzMTMxLCJsb24iOjczLjgwNTQ5OCwicGxhY2VJZCI6IkNoSUpzLXE5ZnplNHdqc1ItS0Nuc2RwbFFpdyIsInBsYWNlTmFtZSI6IlBpbXByaS1DaGluY2h3YWQiLCJzaG93TWFwIjpmYWxzZX1d&radius=2.0&city=pune&locality=Pimpri-Chinchwad&isMetro=false"),
    ("Kalewadi",
     "https://www.nobroker.in/property/commercial/rent/pune/Kalewadi?searchParam=W3sibGF0IjoxOC42MTcxMjE5LCJsb24iOjczLjc5MDI2MDMsInBsYWNlSWQiOiJDaElKcFY3bkhLdTV3anNSY2lYdlZwUGpCTlEiLCJwbGFjZU5hbWUiOiJLYWxld2FkaSIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Kalewadi&isMetro=false"),
    ("Wakad",
     "https://www.nobroker.in/property/commercial/rent/pune/Wakad?searchParam=W3sibGF0IjoxOC42MDEwOTIxLCJsb24iOjczLjc2NDEyNDUsInBsYWNlSWQiOiJDaElKN1J0WHIzcTV3anNSYzJhbnBXczBad3ciLCJwbGFjZU5hbWUiOiJXYWthZCIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Wakad&isMetro=false"),
    ("Viman Nagar",
     "https://www.nobroker.in/property/commercial/rent/pune/Viman%20Nagar?searchParam=W3sibGF0IjoxOC41Njc4NTUzLCJsb24iOjczLjkxNDM2MzcsInBsYWNlSWQiOiJDaElKdFlRVTVrYkJ3anNSc0xPMHFQY3NTTFkiLCJwbGFjZU5hbWUiOiJWaW1hbiBOYWdhciIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Viman%20Nagar&isMetro=false"),
    ("Swargate",
     "https://www.nobroker.in/property/commercial/rent/pune/Swargate?searchParam=W3sibGF0IjoxOC41MDE4MzIyLCJsb24iOjczLjg2MzU5MTIsInBsYWNlSWQiOiJDaElKeFJnWkp4VEF3anNSUDAxSkREX21QUG8iLCJwbGFjZU5hbWUiOiJTd2FyZ2F0ZSIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Swargate&isMetro=false"),
    ("Yerwada",
     "https://www.nobroker.in/property/commercial/rent/pune/Yerawada?searchParam=W3sibGF0IjoxOC41NTI5MjI2LCJsb24iOjczLjg3OTcxMSwicGxhY2VJZCI6IkNoSUpSMER5blNEQndqc1I1V2RrazRkODI0ayIsInBsYWNlTmFtZSI6IlllcmF3YWRhIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Yerawada&isMetro=false"),
    ("Deccan",
     "https://www.nobroker.in/property/commercial/rent/pune/Deccan%20Gymkhana?searchParam=W3sibGF0IjoxOC41MTY1NDM2LCJsb24iOjczLjgzNjYwMjEsInBsYWNlSWQiOiJDaElKZlM4cU1JLV93anNSWkM1YzFyRjJoYTQiLCJwbGFjZU5hbWUiOiJEZWNjYW4gR3lta2hhbmEiLCJzaG93TWFwIjpmYWxzZX1d&radius=2.0&city=pune&locality=Deccan%20Gymkhana&isMetro=false"),
    ("Magarpatta",
     "https://www.nobroker.in/property/commercial/rent/pune/Magarpatta%20Aalaap?searchParam=W3sibGF0IjoxOC40NjAxMjQ2LCJsb24iOjczLjc4NTg0MzcsInBsYWNlSWQiOiI4YTlmOWE5Mjk3ZmRlODk4MDE5N2ZlNWRhODU0Mzk1NF9OQkIiLCJwbGFjZU5hbWUiOiJNYWdhcnBhdHRhIEFhbGFhcCIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Magarpatta%20Aalaap&isMetro=false"),
    ("Koregaon Park",
     "https://www.nobroker.in/property/commercial/rent/pune/Koregaon%20Park?searchParam=W3sibGF0IjoxOC41MzYyMDg0LCJsb24iOjczLjg5Mzk3NDgsInBsYWNlSWQiOiJDaElKZFJJcU9FN0F3anNSOHM4dDRUdnV5QjQiLCJwbGFjZU5hbWUiOiJLb3JlZ2FvbiBQYXJrIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Koregaon%20Park&isMetro=false"),
    ("Pashan",
     "https://www.nobroker.in/property/commercial/rent/pune/Pashan?searchParam=W3sibGF0IjoxOC41NDE0MjY3LCJsb24iOjczLjc5MjM5NTcsInBsYWNlSWQiOiJDaElKYmVSNWktTy13anNSZnRsM1BScV9wWU0iLCJwbGFjZU5hbWUiOiJQYXNoYW4iLCJzaG93TWFwIjpmYWxzZX1d&radius=2.0&city=pune&locality=Pashan&isMetro=false"),
    ("Bavdhan",
     "https://www.nobroker.in/property/commercial/rent/pune/Bavdhan?searchParam=W3sibGF0IjoxOC41MTM0NjMsImxvbiI6NzMuNzY5ODU3ODk5OTk5OTksInBsYWNlSWQiOiJDaElKMDBYVHQxNi13anNSaDVjUmhEeWRFUEEiLCJwbGFjZU5hbWUiOiJCYXZkaGFuIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Bavdhan&isMetro=false"),
    ("Nigdi",
     "https://www.nobroker.in/property/commercial/rent/pune/Nigdi?searchParam=W3sibGF0IjoxOC42NDkxNzM1LCJsb24iOjczLjc3MDY1MzIsInBsYWNlSWQiOiJDaElKdV9ueWJPZTV3anNSeU1hOElkR213bEUiLCJwbGFjZU5hbWUiOiJOaWdkaSIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Nigdi&isMetro=false"),
    ("Camp",
     "https://www.nobroker.in/property/commercial/rent/pune/Camp?searchParam=W3sibGF0IjoxOC41MTIyMzA2LCJsb24iOjczLjg4NjAxLCJwbGFjZUlkIjoiQ2hJSndZa1FuMHpBd2pzUjFKLTQwYW0zeHdvIiwicGxhY2VOYW1lIjoiQ2FtcCIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Camp&isMetro=false"),
    ("Lohegaon",
     "https://www.nobroker.in/property/commercial/rent/pune/Lohegaon?searchParam=W3sibGF0IjoxOC41OTQ0NzA0LCJsb24iOjczLjkyNzU4ODksInBsYWNlSWQiOiJDaElKbWNVLVpxUEd3anNSbHE3YlAzcDJ3ME0iLCJwbGFjZU5hbWUiOiJMb2hlZ2FvbiIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Lohegaon&isMetro=false"),
    ("Katraj",
     "https://www.nobroker.in/property/commercial/rent/pune/Katraj?searchParam=W3sibGF0IjoxOC40NTI5MzIyLCJsb24iOjczLjg2NTIzNzk5OTk5OTk5LCJwbGFjZUlkIjoiQ2hJSmgxOWxBY0hxd2pzUnlKb00xaUFLVzh3IiwicGxhY2VOYW1lIjoiS2F0cmFqIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Katraj&isMetro=false"),
    ("Aundh",
     "https://www.nobroker.in/property/commercial/rent/pune/Aundh?searchParam=W3sibGF0IjoxOC41NjAxNjQ5LCJsb24iOjczLjgwMzEzMzUsInBsYWNlSWQiOiJDaElKelVGZ09raV93anNSTFRyZjJYN2dhTmsiLCJwbGFjZU5hbWUiOiJBdW5kaCIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Aundh&isMetro=false"),
    ("Kondhwa",
     "https://www.nobroker.in/property/commercial/rent/pune/Kondhwa?searchParam=W3sibGF0IjoxOC40Njk1MDg4LCJsb24iOjczLjg4ODk3NzksInBsYWNlSWQiOiJDaElKVC0zeGRZcnF3anNSU3ZvbTlPSVZZTXciLCJwbGFjZU5hbWUiOiJLb25kaHdhIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Kondhwa&isMetro=false"),
    ("Balewadi",
     "https://www.nobroker.in/property/commercial/rent/pune/Balewadi?searchParam=W3sibGF0IjoxOC41Nzg4OTEzLCJsb24iOjczLjc3MDY4MDcsInBsYWNlSWQiOiJDaElKblFrb1FUYTV3anNSWDVZcEdaQXVpVU0iLCJwbGFjZU5hbWUiOiJCYWxld2FkaSIsInNob3dNYXAiOmZhbHNlfV0=&radius=2.0&city=pune&locality=Balewadi&isMetro=false"),
    ("Wagholi",
     "https://www.nobroker.in/property/commercial/rent/pune/Wagholi?searchParam=W3sibGF0IjoxOC41ODA3NzE5LCJsb24iOjczLjk3ODcwNjMsInBsYWNlSWQiOiJDaElKZF9qZW40SER3anNSNEVzMVg1Zy1HZFEiLCJwbGFjZU5hbWUiOiJXYWdob2xpIiwic2hvd01hcCI6ZmFsc2V9XQ==&radius=2.0&city=pune&locality=Wagholi&isMetro=false"),
]

# ---------- Configuration ----------
BASE = "https://www.nobroker.in"
HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/114.0.0.0 Safari/537.36"
    )
}
REQUEST_SLEEP = 0.6
OUTPUT_CSV = "pune_commercial_listings_nobroker.csv"

session = requests.Session()
session.headers.update(HEADERS)

# ---------- Text cleaning (same as before) ----------
def _score_mojibake(s):
    return (s.count("Ã") + s.count("Â") + s.count("â") + s.count("�"))

def clean_text(s):
    if s is None:
        return None
    s = str(s)
    s = s.replace("\u200b", "").replace("\ufeff", "").strip()
    candidates = [s]
    try:
        candidates.append(s.encode("latin1").decode("utf-8"))
    except Exception:
        pass
    try:
        candidates.append(s.encode("cp1252").decode("utf-8"))
    except Exception:
        pass
    try:
        candidates.append(s.encode("utf-8").decode("latin1"))
    except Exception:
        pass
    try:
        temp = s.encode("latin1").decode("utf-8")
        candidates.append(temp.encode("latin1").decode("utf-8"))
    except Exception:
        pass

    best = s
    best_score = _score_mojibake(s)
    for cand in candidates:
        sc = _score_mojibake(cand)
        if "₹" in cand and "₹" in best:
            if sc < best_score:
                best, best_score = cand, sc
        elif "₹" in cand and "₹" not in best:
            best, best_score = cand, sc
        else:
            if sc < best_score:
                best, best_score = cand, sc

    s = best
    repl_map = {
        "Ã¢Â‚Â¹": "₹",
        "â‚¹": "₹",
        "â\x82¹": "₹",
        "\xc2\xa0": " ",
        "\u00a0": " ",
        "Ã¢Â€Â“": "-",
        "Ã¢Â€Âœ": '"',
        "Ã¢Â€Â�": '"',
        "â€”": "-",
        "â€“": "-",
        "â€™": "'",
        "â€œ": '"',
        "â€\x9d": '"',
    }
    for k, v in repl_map.items():
        if k in s:
            s = s.replace(k, v)

    s = re.sub(r"â(?=\s*\d)", "₹", s)
    s = re.sub(r"[ÂÃ]+(?=\s|$)", "", s)
    s = s.replace("�", "")

    try:
        s = unicodedata.normalize("NFKC", s)
    except Exception:
        pass

    s = re.sub(r"[\x00-\x1f\x7f-\x9f]", "", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s if s != "" else None

# ---------- Utility helpers ----------
def safe_text(tag):
    return tag.get_text(" ", strip=True) if tag else None

def parse_amount_rupee(text):
    if not text:
        return None
    t = str(text).replace("\u200b", "").replace(",", "").replace("₹", "").strip()
    if not t:
        return None
    m = re.search(r"([\d.]+)\s*(lacs?|lakh|lakhs|L|k|K)?", t, re.I)
    if not m or not m.group(1):
        return None
    try:
        val = float(m.group(1).strip())
    except Exception:
        return None
    unit = (m.group(2) or "").lower()
    if "lac" in unit or "lakh" in unit or unit == "l":
        return int(val * 100000)
    if unit == "k":
        return int(val * 1000)
    return int(val)

def parse_sqft(text):
    if not text:
        return None
    txt = str(text)
    m = re.search(r"([\d,]+(?:\.\d+)?)\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)\b", txt, re.I)
    if m and m.group(1):
        num = m.group(1).replace(",", "").strip()
        try:
            return float(num)
        except Exception:
            return None
    return None

# ---------- Parking, price-per-sqft, furnishing, sqft, rent/deposit, etc. ----------
_VALID_PARKING_WORDS = {
    "public", "covered", "covered parking", "open", "yes", "no", "no parking",
    "basement", "valet", "private", "street", "common", "none", "open parking"
}
_VALID_PARKING_RE = re.compile(r"^(?:" + r"|".join(re.escape(w) for w in _VALID_PARKING_WORDS) + r")$", re.I)

def nearby_text_candidates(node, max_tokens=8):
    out = []
    parent = getattr(node, "parent", None)
    if parent:
        for child in parent.find_all(recursive=False):
            txt = safe_text(child)
            if txt:
                out.append(txt.strip())
    try:
        prev_sib = node.previous_sibling
        next_sib = node.next_sibling
        for s in (prev_sib, next_sib):
            if s:
                if getattr(s, "get_text", None):
                    txt = safe_text(s)
                    if txt:
                        out.append(txt.strip())
                else:
                    val = str(s).strip()
                    if val:
                        out.append(val)
    except Exception:
        pass
    if parent:
        for s in parent.stripped_strings:
            out.append(s.strip())
            if len(out) >= max_tokens:
                break
    seen = set()
    result = []
    for t in out:
        if not t or t in seen:
            continue
        seen.add(t)
        result.append(t)
    return result[:max_tokens]

def is_good_parking_candidate(s):
    if not s:
        return False
    s = s.strip()
    if re.search(r"\d", s):
        return False
    if "₹" in s or "rs" in s.lower():
        return False
    if "sq" in s.lower() or "per" in s.lower():
        return False
    normalized = re.sub(r"[^A-Za-z\s\-]", "", s).strip().lower()
    if not normalized:
        return False
    if _VALID_PARKING_RE.match(normalized):
        return True
    if len(normalized) <= 22 and re.match(r"^[a-z\s\-]+$", normalized):
        if any(k in normalized for k in ["park", "valet", "basement", "covered", "public", "street", "open", "private", "common", "none"]):
            return True
    return False

def extract_parking(card):
    label_nodes = card.find_all(string=re.compile(r"\bParking\b", re.I))
    for node in label_nodes:
        candidates = nearby_text_candidates(node)
        for cand in candidates:
            if is_good_parking_candidate(cand):
                return cand.strip()
        parent = node.parent
        if parent:
            for attr in ("title", "aria-label"):
                v = parent.get(attr)
                if v and is_good_parking_candidate(v):
                    return v.strip()
    text = " ".join(card.stripped_strings)
    m = re.search(r"\b(Public|Covered|Valet|Basement|Open|Private|Street|Common|No Parking|No)\b", text, re.I)
    if m:
        cand = m.group(1)
        if is_good_parking_candidate(cand):
            return cand.strip()
    return None

def extract_price_per_sqft_from_card(card):
    def get_numeric_before_per_sq(s):
        if not s:
            return None
        s = s.replace("\u200b", " ").strip()
        m = re.search(r"([\d,.]+)\s*(?=(?:per|\b/|\bper\b).*(?:sq|Sq|SQ))", s, re.I)
        if m and m.group(1):
            try:
                return float(m.group(1).replace(",", ""))
            except Exception:
                return None
        m2 = re.search(r"([\d,.]+)", s)
        if m2 and m2.group(1) and "sq" in s.lower():
            try:
                return float(m2.group(1).replace(",", ""))
            except Exception:
                return None
        return None

    for tag in card.find_all(["div", "span"]):
        cl = tag.get("class") or []
        if any("bg-grayDark" in c for c in cl):
            val = safe_text(tag)
            n = get_numeric_before_per_sq(val)
            if n:
                return n
    for tag in card.find_all(["div", "span"]):
        txt = safe_text(tag) or ""
        if "per" in txt.lower() and "sq" in txt.lower():
            n = get_numeric_before_per_sq(txt)
            if n:
                return n
    return None

def extract_furnishing(card):
    for s in card.stripped_strings:
        if re.search(r"\b(Unfurnished|Furnished|Semi[- ]Furnished|Semi[- ]Fitted)\b", s, re.I):
            m = re.search(r"\b(Unfurnished|Furnished|Semi[- ]Furnished|Semi[- ]Fitted)\b", s, re.I)
            if m:
                return m.group(1)
    nodes = card.find_all(string=re.compile(r"\bFurnishing\b", re.I))
    for node in nodes:
        parent = node.parent
        if parent:
            candidate = parent.find(lambda t: t.name in ("div", "span")
                                              and t.get("class")
                                              and "font-semi-bold" in " ".join(t.get("class")))
            if candidate:
                v = safe_text(candidate) or ""
                if "₹" in v or "sq" in v.lower() or re.search(r"\d{3,}", v):
                    continue
                if re.match(r"^[A-Za-z\s\-]{2,30}$", v.strip()):
                    return v.strip()
    return None

def extract_sqft_near_label(card):
    label = card.find(lambda t: t.name in ("div", "span")
                               and re.search(r"Built[-\s]?up|Builtup|Built Up|Builtup",
                                             (t.get_text(" ", strip=True) or ""), re.I))
    if label:
        parent = label.parent
        if parent:
            val = parent.find(lambda t: t.name in ("div", "span")
                                        and t.get("class")
                                        and "font-semi-bold" in " ".join(t.get("class")))
            if val:
                txt = safe_text(val) or ""
                if re.search(r"\d.+(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)", txt, re.I):
                    return txt
            ptxt = safe_text(parent) or ""
            m = re.search(r"(\d{2,6}(?:\.\d+)?)\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)", ptxt or "", re.I)
            if m:
                return m.group(0)
    for t in card.find_all(["div", "span", "p", "li"]):
        txt = safe_text(t) or ""
        if re.search(r"\d{2,6}(?:\.\d+)?\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)", txt, re.I):
            return re.search(r"\d{2,6}(?:\.\d+)?\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m|sq m)", txt, re.I).group(0)
    return None

def extract_rent_from_card(card):
    for s in card.stripped_strings:
        txt = s.strip()
        if "₹" in txt:
            return txt
    vals = [safe_text(x) for x in card.find_all(lambda t: t.name == "div"
                                                          and t.get("class")
                                                          and "font-semi-bold" in " ".join(t.get("class")))]
    for v in vals:
        if v and ("₹" in v or re.search(r"\d", v)):
            return v
    return None

def extract_deposit(card):
    dep = None
    nodes = card.find_all(string=re.compile(r"\bDeposit\b", re.I))
    for node in nodes:
        parent = node.parent
        if parent:
            candidate = parent.find(lambda t: t.name in ("div", "span")
                                              and t.get("class")
                                              and "font-semi-bold" in " ".join(t.get("class")))
            if candidate:
                dep = safe_text(candidate)
                break
    if dep:
        return dep
    vals = [safe_text(x) for x in card.find_all(lambda t: t.name == "div"
                                                          and t.get("class")
                                                          and "font-semi-bold" in " ".join(t.get("class")))]
    if len(vals) >= 2:
        return vals[1]
    return None

def extract_available_from(card):
    for s in card.stripped_strings:
        txt = s.strip()
        if re.match(r"\d{4}-\d{2}-\d{2}", txt):
            return txt
    nodes = card.find_all(string=re.compile(r"Available\s*From|Available from|Available\b", re.I))
    for node in nodes:
        parent = node.parent
        if parent:
            candidate = parent.find(lambda t: t.name in ("div", "span")
                                              and t.get("class")
                                              and "font-semi-bold" in " ".join(t.get("class")))
            if candidate:
                return safe_text(candidate)
    return None

def extract_property_type(card):
    txt = " ".join(card.stripped_strings)
    m = re.search(r"\b(Shop|Office Space|Office|Retail|Warehouse|Showroom|Commercial|Restaurant)\b", txt, re.I)
    return m.group(1) if m else None

# ---------- Detail page parser (unchanged) ----------
def parse_detail_page(detail_url):
    if not detail_url:
        return {}
    try:
        r = session.get(detail_url, timeout=12)
        time.sleep(REQUEST_SLEEP)
    except Exception:
        return {}
    if r.status_code != 200:
        return {}
    dsoup = BeautifulSoup(r.text, "html.parser")
    out = {}
    text = dsoup.get_text(" ", strip=True)
    for tag in dsoup.find_all("script", {"type": "application/ld+json"}):
        try:
            import json
            obj = json.loads(tag.string or "{}")
            if isinstance(obj, dict):
                if "price" in obj and not out.get("Rent_detail"):
                    out["Rent_detail"] = obj.get("price")
                if "address" in obj and not out.get("Location_detail"):
                    addr = obj.get("address")
                    if isinstance(addr, dict):
                        out["Location_detail"] = addr.get("streetAddress") or addr.get("addressLocality")
                    elif isinstance(addr, str):
                        out["Location_detail"] = addr
        except Exception:
            pass
    m_sq = re.search(r"([\d,]+(?:\.\d+)?)\s*(sq\.?ft|sqft|sq ft|sqm|m2|sq\.?m)", text, re.I)
    if m_sq:
        out["Sqft_detail"] = m_sq.group(0)
    m_dep = re.search(r"([₹\d,.\sKkLl]+)\s*(?:Deposit|deposit)", text)
    if m_dep:
        out["Deposit_detail"] = m_dep.group(1)
    m_furn = re.search(r"\b(Unfurnished|Furnished|Semi[- ]Furnished)\b", text, re.I)
    if m_furn:
        out["Furnishing_detail"] = m_furn.group(1)
    m_av = re.search(r"Available\s*From[:\s]*([A-Za-z0-9\-\s]+)", text, re.I)
    if m_av:
        out["Available_from_detail"] = m_av.group(1).strip()
    m_park = re.search(r"\b(Public|Covered|No Parking|Yes|No|Basement|Valet|Private)\b\s*(?:Parking)", text, re.I)
    if m_park:
        out["Parking_detail"] = m_park.group(1)
    return out

# ---------- Main page parsing ----------
def parse_search_page(html, locality_name, follow_details=False, detail_limit=None):
    soup = BeautifulSoup(html, "html.parser")
    cards = soup.find_all("article")
    rows = []
    for idx, card in enumerate(cards, start=1):
        title_tag = card.find("h2", attrs={"class": lambda c: c and "heading-6" in c})
        title = safe_text(title_tag) or safe_text(card.find(["h2", "h3", "h4"]))

        loc_tag = card.find(
            lambda t: t.name in ("div", "span")
            and t.get("class")
            and "overflow-ellipsis" in " ".join(t.get("class"))
            and "whitespace-nowrap" in " ".join(t.get("class"))
        )
        location = safe_text(loc_tag) if loc_tag else None
        if location and title and title.strip() in location:
            location = location.replace(title, "").strip(" ,-")

        rent_raw = extract_rent_from_card(card)
        deposit_raw = extract_deposit(card)
        sqft_raw = extract_sqft_near_label(card)
        sqft_val = parse_sqft(sqft_raw) if sqft_raw else None

        price_pps = extract_price_per_sqft_from_card(card)
        rent_inr = parse_amount_rupee(rent_raw) if rent_raw else None
        if not price_pps and rent_inr and sqft_val and sqft_val > 0:
            try:
                price_pps = round(rent_inr / sqft_val, 2)
            except Exception:
                price_pps = None

        parking = extract_parking(card)
        furnishing = extract_furnishing(card)
        available_from = extract_available_from(card)
        property_type = extract_property_type(card)

        a = card.find("a", href=True)
        detail_link = urljoin(BASE, a.get("href")) if a else None

        detail_data = {}
        if follow_details:
            if detail_limit is not None and idx > detail_limit:
                detail_data = {}
            else:
                detail_data = parse_detail_page(detail_link)

        if detail_data.get("Location_detail"):
            location = detail_data.get("Location_detail")
        if detail_data.get("Rent_detail"):
            rent_raw = detail_data.get("Rent_detail")
            rent_inr = parse_amount_rupee(rent_raw)
        if detail_data.get("Deposit_detail"):
            deposit_raw = detail_data.get("Deposit_detail")
        if detail_data.get("Sqft_detail"):
            sqft_raw = detail_data.get("Sqft_detail")
            sqft_val = parse_sqft(sqft_raw)
        if detail_data.get("Furnishing_detail"):
            furnishing = detail_data.get("Furnishing_detail")
        if detail_data.get("Available_from_detail"):
            available_from = detail_data.get("Available_from_detail")
        if detail_data.get("Parking_detail"):
            parking = detail_data.get("Parking_detail")

        title_c = clean_text(title)
        location_c = clean_text(location)
        rent_raw_c = clean_text(rent_raw)
        deposit_raw_c = clean_text(deposit_raw)
        sqft_raw_c = clean_text(sqft_raw)
        parking_c = clean_text(parking)
        furnishing_c = clean_text(furnishing)
        available_from_c = clean_text(available_from)
        property_type_c = clean_text(property_type)

        row = {
            "Locality": locality_name,
            "Title": title_c,
            "Location": location_c,
            "Rent_raw": rent_raw_c,
            "Rent_inr": rent_inr,
            "Deposit_raw": deposit_raw_c,
            "Sqft_raw": sqft_raw_c,
            "Sqft": sqft_val,
            "Price_per_sqft": price_pps,
            "Parking": parking_c,
            "Furnishing": furnishing_c,
            "Available_from": available_from_c,
            "Property_type": property_type_c,
            "Detail_URL": detail_link,
        }
        rows.append(row)
    return rows

# ---------- Runner for a single locality ----------
def run_for_locality(locality_name, search_url, follow_details=False, detail_limit=None):
    print(f"Requesting search page for locality: {locality_name} -> {search_url}")
    r = session.get(search_url, timeout=15)
    r.raise_for_status()
    time.sleep(REQUEST_SLEEP)

    rows = parse_search_page(
        r.text,
        locality_name=locality_name,
        follow_details=follow_details,
        detail_limit=detail_limit,
    )
    df = pd.DataFrame(rows)

    for col in ["Deposit_raw", "Rent_raw", "Location", "Title", "Parking", "Furnishing", "Sqft_raw"]:
        if col in df.columns:
            df[col] = df[col].astype(str).apply(clean_text)

    print(f"Locality '{locality_name}' — rows scraped: {len(df)}")
    return df

# ---------- Main: loop all localities and append to same CSV ----------
if __name__ == "__main__":
    all_dfs = []

    for loc_name, loc_url in locality_links:
        try:
            df_loc = run_for_locality(loc_name, loc_url, follow_details=False, detail_limit=None)
            if not df_loc.empty:
                all_dfs.append(df_loc)
        except Exception as e:
            print(f"Error while scraping locality '{loc_name}': {e}")

    if all_dfs:
        combined = pd.concat(all_dfs, ignore_index=True)

        # IMPORTANT: close the CSV in Excel / other apps before running to avoid PermissionError
        if os.path.exists(OUTPUT_CSV):
            combined.to_csv(
                OUTPUT_CSV,
                mode="a",
                index=False,
                header=False,  # no header on append
                encoding="utf-8",
            )
            print(f"Appended {len(combined)} rows to existing {OUTPUT_CSV}")
        else:
            combined.to_csv(
                OUTPUT_CSV,
                index=False,
                encoding="utf-8-sig",  # BOM for Excel
            )
            print(f"Created {OUTPUT_CSV} with {len(combined)} rows")

        print(combined.info())
        print(combined.head(10).to_string(index=False))
    else:
        print("No data scraped for any locality.")
```
